```python
# 在已有的导入语句中添加这一行
import numpy as np  # 导入numpy并简写为np

# 其他已有导入（保持不变）
import import_ipynb
import os
import torch
import torch.optim as optim
import torch.nn as nn
from torch.utils.data import TensorDataset, DataLoader
import matplotlib.pyplot as plt
from data_loader import (
    load_images_from_dir, add_underwater_noise, preprocess_data
)
from model import DenoiseAE

# --------------------------
# 随机种子设置（确保结果可复现）
# 若需要复现报告中的结果，保持 seed=42 不变；
# 若想尝试不同初始化效果，可修改 seed 数值（如 seed=100）
# --------------------------
import random
import numpy as np
import torch

def set_seed(seed=42):
    random.seed(seed)  # Python基础随机种子
    np.random.seed(seed)  # numpy随机种子（控制噪声生成、数据打乱）
    torch.manual_seed(seed)  # PyTorch CPU随机种子（控制模型初始化）
    if torch.cuda.is_available():
        torch.cuda.manual_seed(seed)  # PyTorch GPU随机种子
        torch.cuda.manual_seed_all(seed)  # 多GPU场景
        torch.backends.cudnn.deterministic = True  # 固定CUDA卷积算法
        torch.backends.cudnn.benchmark = False  # 关闭自动选择最快算法（避免随机性）

# 调用函数固定种子
set_seed(seed=42)

# 1. 配置参数
TRAIN_IMG_DIR = r"C:\Users\27976\Desktop\UATD\UATD_Training\images"  # 替换为你的训练集路径
TEST_IMG_DIR = r"C:\Users\27976\Desktop\UATD\UATD_Test_1\images"      # 替换为你的测试集路径
NUM_TRAIN = 500       # 训练样本数
NUM_TEST = 100        # 测试样本数
IMAGE_SIZE = (256, 256)  # 图像尺寸
BATCH_SIZE = 16       # 批次大小
EPOCHS = 50           # 训练轮次
LEARNING_RATE = 0.001 # 学习率
SAVE_PATH = "denoise_ae.pth"  # 模型保存路径
RESULTS_DIR = "results"       # 结果图像保存文件夹

# 确保结果文件夹存在
os.makedirs(RESULTS_DIR, exist_ok=True)

# 2. 加载数据
print("加载数据中...")
train_clean = load_images_from_dir(TRAIN_IMG_DIR, num_samples=NUM_TRAIN, size=IMAGE_SIZE)
test_clean = load_images_from_dir(TEST_IMG_DIR, num_samples=NUM_TEST, size=IMAGE_SIZE)
train_noisy = np.array([add_underwater_noise(img) for img in train_clean])
test_noisy = np.array([add_underwater_noise(img) for img in test_clean])

# 预处理为Tensor
train_clean_tensor, train_noisy_tensor = preprocess_data(train_clean, train_noisy)
test_clean_tensor, test_noisy_tensor = preprocess_data(test_clean, test_noisy)

# 创建DataLoader
train_dataset = TensorDataset(train_noisy_tensor, train_clean_tensor)
test_dataset = TensorDataset(test_noisy_tensor, test_clean_tensor)
train_loader = DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=BATCH_SIZE, shuffle=False)

# 3. 初始化模型、损失函数、优化器
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = DenoiseAE().to(device)
criterion = nn.MSELoss()  # 均方误差损失（适合图像重构）
optimizer = optim.Adam(model.parameters(), lr=LEARNING_RATE)

# 4. 训练模型
print(f"使用{device}训练，共{EPOCHS}轮...")
train_losses = []
test_losses = []

for epoch in range(EPOCHS):
    # 训练阶段
    model.train()
    train_loss = 0.0
    for noisy_imgs, clean_imgs in train_loader:
        noisy_imgs, clean_imgs = noisy_imgs.to(device), clean_imgs.to(device)
        
        optimizer.zero_grad()  # 清空梯度
        outputs = model(noisy_imgs)  # 前向传播
        loss = criterion(outputs, clean_imgs)  # 计算损失
        loss.backward()  # 反向传播
        optimizer.step()  # 更新参数
        
        train_loss += loss.item() * noisy_imgs.size(0)  # 累计损失
    
    # 计算平均训练损失
    train_loss_avg = train_loss / len(train_loader.dataset)
    train_losses.append(train_loss_avg)
    
    # 测试阶段（无梯度计算）
    model.eval()
    test_loss = 0.0
    with torch.no_grad():
        for noisy_imgs, clean_imgs in test_loader:
            noisy_imgs, clean_imgs = noisy_imgs.to(device), clean_imgs.to(device)
            outputs = model(noisy_imgs)
            loss = criterion(outputs, clean_imgs)
            test_loss += loss.item() * noisy_imgs.size(0)
    
    test_loss_avg = test_loss / len(test_loader.dataset)
    test_losses.append(test_loss_avg)
    
    # 打印进度
    print(f"Epoch {epoch+1:2d}/{EPOCHS}, 训练损失: {train_loss_avg:.6f}, 测试损失: {test_loss_avg:.6f}")

# 5. 保存模型
torch.save(model.state_dict(), SAVE_PATH)
print(f"模型已保存至：{SAVE_PATH}")

# 6. 绘制并保存损失曲线
plt.figure(figsize=(10, 5))
plt.plot(train_losses, label="训练损失")
plt.plot(test_losses, label="测试损失")
plt.xlabel("训练轮次（Epoch）")
plt.ylabel("MSE损失")
plt.title("模型训练损失曲线")
plt.legend()
loss_curve_path = os.path.join(RESULTS_DIR, "loss_curve.png")
plt.savefig(loss_curve_path)
plt.close()
print(f"损失曲线已保存至：{loss_curve_path}")
```
